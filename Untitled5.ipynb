```python
import torch
from torch.nn import Linear, MSELoss
from torch.optim import Adam
from tqdm import trange
import matplotlib.pyplot as plt

# 1. 定义输入与目标
five = torch.tensor([5.0])
three = torch.tensor([3.0])
eight = torch.tensor([8.0])

# 合并输入 [5,3] → [1,2]
input = torch.cat([five, three]).unsqueeze(0)   # shape [1,2]
target = eight.unsqueeze(0)                     # shape [1,1]

# 2. 定义模型 f
f = Linear(2, 1)   # 输入2维，输出1维

# 3. 定义优化器与损失函数
optimizer = Adam(f.parameters(), lr=1e-4)
mse = MSELoss()

# 4. 训练循环
pbar = trange(15000)
outputs = []

for i in pbar:
    output = f(input)            # 前向传播
    loss = mse(output, target)   # 计算损失

    optimizer.zero_grad()        # 梯度清零
    loss.backward()              # 反向传播
    optimizer.step()             # 参数更新

    pbar.set_description(f'output = {output.item():.4f}')

    if i % 100 == 0:
        outputs.append(output.item())

# 5. 可视化结果
plt.plot(outputs)
plt.xlabel("Iterations (x100)")
plt.ylabel("Output value")
plt.title("Training Progress")
plt.show()

# 6. 测试：输入 [2,4]
test = torch.tensor([2.0, 4.0]).unsqueeze(0)
print("Test input [2,4] →", f(test).item())
```


说明（中文）
这一段代码构成了模型的训练循环。首先定义 Adam 优化器和 MSELoss 损失函数，用于更新参数和度量误差。在 15,000 次迭代中，模型通过前向传播得到输出 output，并与目标值 eight 计算误差。之后通过反向传播更新参数，不断减小误差。程序每 100 次保存一次 output 以便绘制曲线。训练结果显示，输出逐渐逼近 8，说明模型已经成功学会了加法近似。

설명문 (한국어)
이 코드는 모델의 학습 루프를 구성한다. 먼저 Adam 옵티마이저와 MSELoss 손실 함수를 정의하여 파라미터 갱신과 오차 측정에 사용한다. 15,000번 반복 동안 모델은 순전파로 output을 계산하고, 목표 값 eight과의 오차를 구한다. 이후 역전파를 통해 파라미터를 갱신하며 오차를 점점 줄여 나간다. 프로그램은 100번마다 output을 저장하여 그래프를 그릴 수 있도록 한다. 학습 결과 출력은 점차 8에 가까워지며, 모델이 덧셈을 근사적으로 학습했음을 보여준다.




```python
from tqdm import trange
pbar = range(50)
outputs = []

for i in pbar:
    output = f(input)
    loss = mse(output, eight)
    print(loss)  # 손실 출력

    optimizer.zero_grad()

    for param in f.parameters():
        print(param.grad)  # 역전파 이전 기울기(grad)

    loss.backward()

    for param in f.parameters():
        print(param.grad)  # 역전파 이후 기울기(grad)

    optimizer.step()

    if i % 100 == 0:
        outputs.append(output.item())
```

说明（中文）
本代码在训练过程中输出了 loss 和 梯度 (gradients)。

每次迭代首先打印当前损失值。

在 loss.backward() 之前，参数的梯度为空。

调用 loss.backward() 后，计算出损失对参数的梯度，并打印出来。

执行 optimizer.step() 后，参数根据梯度进行更新，从而逐步降低损失。

通过这种方式，可以直观地观察 梯度下降 的过程，加深对深度学习优化机制的理解。

설명문 (한국어)
이 코드는 학습 과정에서 손실(loss) 과 기울기(gradients) 를 출력한다.

매 반복(iteration)마다 현재 손실 값을 출력한다.

loss.backward() 이전에는 파라미터의 기울기가 비어 있다.

loss.backward()를 호출하면 손실에 대한 파라미터의 기울기가 계산되어 출력된다.

optimizer.step()을 실행하면 기울기에 따라 파라미터가 갱신되고, 손실이 점차 줄어든다.

이 과정을 통해 경사 하강법(gradient descent) 의 원리를 직접 확인할 수 있으며, 딥러닝 최적화 메커니즘에 대한 이해를 높일 수 있다.

```python
import torch
from torch.nn import Linear, MSELoss
from torch.optim import Adam
from tqdm import trange
import matplotlib.pyplot as plt

# 自动检测设备 (GPU 优先，否则 CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# 定义输入与目标
five = torch.tensor([5.0], device=device)
three = torch.tensor([3.0], device=device)
eight = torch.tensor([8.0], device=device)

# 输入 [5,3] → [1,2]
input = torch.cat([five, three]).unsqueeze(0)
target = eight.unsqueeze(0)

# 定义模型 (2 输入 → 1 输出)
f = Linear(2, 1).to(device)

# 定义优化器和损失函数（学习率调大，加快收敛）
optimizer = Adam(f.parameters(), lr=1e-2)
mse = MSELoss()

# 训练循环（快速版：只跑 2000 步）
pbar = trange(2000)
outputs = []

for i in pbar:
    output = f(input)
    loss = mse(output, target)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if i % 50 == 0:   # 每 50 步记录一次结果
        outputs.append(output.item())
        pbar.set_description(f'output = {output.item():.4f}')

# 可视化结果
plt.plot(outputs)
plt.xlabel("Iterations (x50)")
plt.ylabel("Output value")
plt.title("Training Progress (Quick Version)")
plt.show()

# 测试
test = torch.tensor([2.0, 4.0], device=device).unsqueeze(0)
print("Test input [2,4] →", f(test).item())
```

说明（中文）
这一实验采用了更快的训练设置：

学习率从 1e-4 提高到 1e-2，加快了收敛速度。

迭代次数减少到 2000，节省运行时间。

每 50 步记录一次结果，用于绘制收敛曲线。

结果显示，模型在较短时间内就能把输出值逐渐逼近目标 8，并且在测试输入 [2,4] 时输出接近 6，验证了模型的加法近似能力。

설명문 (한국어)
이번 실험은 더 빠른 학습 설정을 사용하였다:

학습률을 1e-4에서 1e-2로 높여 수렴 속도를 빠르게 하였다.

반복 횟수를 2000으로 줄여 실행 시간을 절약하였다.

50번마다 출력을 기록하여 수렴 곡선을 확인할 수 있도록 하였다.

그 결과, 모델은 짧은 시간 안에 출력 값을 목표 8에 근접시킬 수 있었으며, 테스트 입력 [2,4]에서도 약 6의 출력을 내어 덧셈 근사 기능을 학습했음을 확인할 수 있었다.

```python
# 单个样本
sample0_input = torch.Tensor([5, 3])
sample0_output = torch.Tensor([8])

sample1_input = torch.Tensor([1, 2])
sample1_output = torch.Tensor([3])

# 多个样本拼接
samples_input = torch.stack([sample0_input, sample1_input])
print(samples_input)

samples_output = torch.stack([sample0_output, sample1_output])
print(samples_output)
```

说明（中文）
这段代码演示了如何在 PyTorch 中处理 多个训练样本：

首先定义了两个样本：[5,3] → 8 和 [1,2] → 3。

使用 torch.stack 将它们拼接成批量输入：

输入 samples_input 形状为 [2,2]，包含两个样本。

输出 samples_output 形状为 [2,1]，对应两个目标值。

这种方式可以一次性输入多个样本进行训练，而不是逐个样本输入，从而更高效地利用批处理。

설명문 (한국어)
이 코드는 PyTorch에서 여러 학습 샘플을 다루는 방법을 보여준다:

먼저 두 개의 샘플을 정의하였다: [5,3] → 8 과 [1,2] → 3.

torch.stack을 사용하여 이들을 하나의 배치(batch)로 합쳤다:

입력 samples_input의 형태는 [2,2]로 두 샘플을 포함한다.

출력 samples_output의 형태는 [2,1]로 두 목표 값을 포함한다.

이 방식은 샘플을 하나씩 처리하지 않고, 배치 학습(batch training) 으로 효율적으로 모델을 학습시킬 수 있도록 해준다.

```python
from tqdm import trange

pbar = range(50)
outputs0 = []
outputs1 = []

for i in pbar:
    output = f(samples_input)                     # 多样本输入
    loss = mse(output, samples_output)            # 损失函数
    print(loss)

    optimizer.zero_grad()

    for param in f.parameters():
        print(param.grad)                         # 反向传播前梯度

    loss.backward()

    for param in f.parameters():
        print(param.grad)                         # 反向传播后梯度

    optimizer.step()

    # 记录两个样本的输出
    outputs0.append(output[0].item())
    outputs1.append(output[1].item())
```

```python
samples_input = torch.stack([torch.Tensor([5,3]), torch.Tensor([1,2])])
samples_output = torch.stack([torch.Tensor([8]), torch.Tensor([3])])

for i in range(500):
    output = f(samples_input)
    loss = mse(output, samples_output)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
from matplotlib import pyplot as plt

plt.plot(outputs0,'r')
plt.plot(outputs1,'b')
plt.show
```

```python
import torch
from torch.nn import Linear, MSELoss
from torch.optim import Adam
import matplotlib.pyplot as plt

# 定义模型、优化器和损失函数
f = Linear(2, 1)
optimizer = Adam(f.parameters(), lr=1e-2)
mse = MSELoss()

# 定义多个样本
samples_input = torch.stack([torch.Tensor([5,3]), torch.Tensor([1,2])])
samples_output = torch.stack([torch.Tensor([8]), torch.Tensor([3])])

# 训练循环，记录两个样本的预测值
outputs0, outputs1 = [], []
for i in range(500):
    output = f(samples_input)
    loss = mse(output, samples_output)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    outputs0.append(output[0].item())
    outputs1.append(output[1].item())

# 可视化两个样本的训练收敛过程
plt.plot(outputs0, 'r', label="[5,3] → 8")
plt.plot(outputs1, 'b', label="[1,2] → 3")
plt.title("Training Progress for Multiple Samples")
plt.xlabel("Iterations")
plt.ylabel("Output Value")
plt.legend()
plt.show()

# 测试新输入
test_input = torch.Tensor([2,4])
print("Test input [2,4] →", f(test_input).item())
```
