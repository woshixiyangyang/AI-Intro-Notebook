```python
import random
import itertools

"""
Flow는 하나의 App (유튜브 비디오 전송, 메세지 전송 등)에 필요한 Packet의 묶음. 즉 플로우의 모든 패킷이 전송 되어야 하나의 유튜브 비디오가 무사히 전송 완료라고 보면 됨.
Packet은 데이터를 전송할때의 가장 작은 유닛.
"""


# 패킷 Class
class Flow:
    def __init__(self, name, packets, arrival_times):
        assert len(packets) == len(arrival_times)
        self.name = name # Flow 이름 --> 큰 상관 없음

        # 버퍼 큐: (고유 id, 도착시각, 패킷크기) 튜플을 저장
        # Note: 각 플로우 마다 큐를 가지고 있음!
        self.queue = []     # 예: [(17, 0.5, 640), (18, 0.7, 512), ...]
        self.all_packets = list(zip(arrival_times, packets))
        self.wait_sum = 0.0 # 각 플로우의 대기 시간
        self.sent_pkts = 0 # 전송 완료된 패킷 수
        self.sent_bytes = 0 # 전송 완료된 바이트 수
        self.dropped = 0  # 버퍼 초과로 드롭된 패킷 수

    def has_packets(self):
        return bool(self.queue) or bool(self.all_packets) # 패킷이 대기 중이거나 도착 예정 패킷이 있으면 True


# Helper: 현재 까지 (now) 도착한 패킷을 각 플로우의 버퍼에 추가
# Note: 각 플로우 마다 큐를 가지고 있음!
# 패킷 추가시 버퍼 사이즈가 이미 꽉 차있다면 패킷은 드롭(Drop)
def receive_arrivals(flows, now, bufsize, id_counter):
    arrivals_flat = []
    for f in flows:
        ready = [p for p in f.all_packets if p[0] <= now]
        f.all_packets = [p for p in f.all_packets if p[0] > now]
        for t_arr, size in ready:
            if len(f.queue) < bufsize:
                pid = next(id_counter)
                f.queue.append((pid, t_arr, size))
                arrivals_flat.append((pid, f.name, t_arr, size))
            else:
                f.dropped += 1
    return arrivals_flat # 이번 시간 (now)에서 도착한 모든 패킷의 리스트.


# Helper: 그 다음 이벤트 발생 시간을 찾음.
# 이벤트 발생은 packet이 도착하는 시간
def next_arrival_time(flows):
    mins = [p[0] for f in flows for p in f.all_packets]
    return min(mins) if mins else None



# 1) Round Robin
"""
아래 코드 작동 원리
    1. 여러 개의 Flow가 있을 때, 각 Flow의 버퍼(queue)에서
        한 번에 하나의 패킷씩 돌아가며 전송하는 방식.
    2. 즉, Flow A -> Flow B -> Flow C-> 다시 Flow A 순서로 반복.
    3. 각 Flow에 동일한 기회를 주기 때문에 '공평한(fair)'한 스케줄링.
"""
def round_robin(flows, bufsize=5, cap=1000, idle=0.1):
    now = 0.0
    tl = []
    id_counter = itertools.count(1)

    while any(f.has_packets() for f in flows):
        receive_arrivals(flows, now, bufsize, id_counter)

        progressed = False
        for f in flows:
            if f.queue:
                pid, t_arr, size = f.queue.pop(0)
                tx = size / cap
                now += tx
                f.sent_pkts += 1
                f.sent_bytes += size
                f.wait_sum += now - t_arr
                tl.append((f.name, size, round(now, 4)))
                receive_arrivals(flows, now, bufsize, id_counter)
                progressed = True

        if not progressed:
            na = next_arrival_time(flows)
            if na is not None and na > now:
                now = na
            else:
                now += idle
    return tl


# 2) FCFS
"""
아래 코드 작동 원리
    1. 모든 Flow에서 도착한 패킷들을 하나의 전역 큐(merged)에 합쳐서,
       "가장 먼저 도착한 패킷"부터 순서대로 전송하는 방식.
    2. 즉, 패킷의 출처(Flow A/B/C)는 중요하지 않음 — 도착 시간이 빠른 순이 우선.
"""
def fcfs(flows, bufsize=5, cap=1000, idle=0.1):
    now = 0.0
    tl = []
    merged = []
    id_counter = itertools.count(1)

    while any(f.has_packets() for f in flows) or merged:
        merged += receive_arrivals(flows, now, bufsize, id_counter)
        if merged:
            pid, name, t_arr, size = merged.pop(0)

            for f in flows:
                if f.name == name:
                    for i, (qa_id, qa_t, qa_s) in enumerate(f.queue):
                        if qa_id == pid:
                            f.queue.pop(i)
                            break
                    tx = size / cap
                    now += tx
                    f.sent_pkts += 1
                    f.sent_bytes += size
                    f.wait_sum += now - t_arr
                    tl.append((name, size, round(now, 4)))
                    merged += receive_arrivals(flows, now, bufsize, id_counter)
                    break
        else:
            na = next_arrival_time(flows)
            if na is not None and na > now:
                now = na
            else:
                now += idle
    return tl



# Main function
if __name__ == "__main__":
    random.seed(42)

    # Simulation Parameters
    NUM_PKTS = 8 # 플로우 당 패킷 수
    BUF = 5 # 각 플로우 버퍼 크기 (버퍼가 담을 수 있는 총 패킷 수)
    CAP = 1000   # 링크 용량 - 데이터를 초당 얼마나 보낼 수 있는지. (bytes per time  unit)

    def make_flow(name, n, lo, hi, gap):
        sizes = [random.randint(lo, hi) for _ in range(n)] # 패킷은 사이즈는 균일하게 분포(uniform distribution) 한다고 가정
        arrivals = [i * gap for i in range(n)]  # 패킷은 균일하게 도착한다고 가정
        return Flow(name, sizes, arrivals)


    # 예제 플로우 생성
    flows_rr = [
        make_flow("A", NUM_PKTS, 200, 800, 0.5),
        make_flow("B", NUM_PKTS, 300, 700, 0.7),
        make_flow("C", NUM_PKTS, 250, 900, 0.9),
    ]
    flows_fc = [
        make_flow("A", NUM_PKTS, 200, 800, 0.5),
        make_flow("B", NUM_PKTS, 300, 700, 0.7),
        make_flow("C", NUM_PKTS, 250, 900, 0.9),
    ]


    # -------- RR 실행 --------
    print("=== Round Robin ===")
    tl_rr = round_robin(flows_rr, bufsize=BUF, cap=CAP)
    for e in tl_rr:
        print(e)

    print("\n--- RR Summary ---")
    for f in flows_rr:
        avg = f.wait_sum / f.sent_pkts if f.sent_pkts else 0.0
        print(f"{f.name}: pkts={f.sent_pkts}, bytes={f.sent_bytes}, drop={f.dropped}, avgDelay={avg:.4f}")

    print("\n" + "="*70 + "\n")


   # -------- FCFS 실행 --------
    print("=== FCFS ===")
    tl_fc = fcfs(flows_fc, bufsize=BUF, cap=CAP)
    for e in tl_fc:
        print(e)

    print("\n--- FCFS Summary ---")
    for f in flows_fc:
        avg = f.wait_sum / f.sent_pkts if f.sent_pkts else 0.0
        print(f"{f.name}: pkts={f.sent_pkts}, bytes={f.sent_bytes}, drop={f.dropped}, avgDelay={avg:.4f}")
```

```python
import random
import itertools
import csv


class Flow:
    def __init__(self, name, packets, arrival_times):
        assert len(packets) == len(arrival_times)
        self.name = name
        self.queue = []
        self.all_packets = list(zip(arrival_times, packets))
        self.wait_sum = 0.0
        self.sent_pkts = 0
        self.sent_bytes = 0
        self.dropped = 0

    def has_packets(self):
        return bool(self.queue) or bool(self.all_packets)


def receive_arrivals(flows, now, bufsize, id_counter):
    arrivals_flat = []
    for f in flows:
        ready = [p for p in f.all_packets if p[0] <= now]
        f.all_packets = [p for p in f.all_packets if p[0] > now]
        for t_arr, size in ready:
            if len(f.queue) < bufsize:
                pid = next(id_counter)
                f.queue.append((pid, t_arr, size))
                arrivals_flat.append((pid, f.name, t_arr, size))
            else:
                f.dropped += 1
    return arrivals_flat


def next_arrival_time(flows):
    mins = [p[0] for f in flows for p in f.all_packets]
    return min(mins) if mins else None


def round_robin(flows, bufsize=5, cap=1000, idle=0.1):
    now = 0.0
    tl = []
    id_counter = itertools.count(1)

    while any(f.has_packets() for f in flows):
        receive_arrivals(flows, now, bufsize, id_counter)
        progressed = False
        for f in flows:
            if f.queue:
                pid, t_arr, size = f.queue.pop(0)
                tx = size / cap
                now += tx
                f.sent_pkts += 1
                f.sent_bytes += size
                f.wait_sum += now - t_arr
                tl.append((f.name, size, round(now, 4)))
                receive_arrivals(flows, now, bufsize, id_counter)
                progressed = True
        if not progressed:
            na = next_arrival_time(flows)
            if na is not None and na > now:
                now = na
            else:
                now += idle
    return tl


def fcfs(flows, bufsize=5, cap=1000, idle=0.1):
    now = 0.0
    tl = []
    merged = []
    id_counter = itertools.count(1)

    while any(f.has_packets() for f in flows) or merged:
        merged += receive_arrivals(flows, now, bufsize, id_counter)
        if merged:
            pid, name, t_arr, size = merged.pop(0)
            for f in flows:
                if f.name == name:
                    for i, (qa_id, qa_t, qa_s) in enumerate(f.queue):
                        if qa_id == pid:
                            f.queue.pop(i)
                            break
                    tx = size / cap
                    now += tx
                    f.sent_pkts += 1
                    f.sent_bytes += size
                    f.wait_sum += now - t_arr
                    tl.append((name, size, round(now, 4)))
                    merged += receive_arrivals(flows, now, bufsize, id_counter)
                    break
        else:
            na = next_arrival_time(flows)
            if na is not None and na > now:
                now = na
            else:
                now += idle
    return tl


def shortest_backlog_first(flows, bufsize=5, cap=1000, idle=0.1):
    """
    Shortest-Backlog-First (SBF):
    每次发送时,选择当前 buffer 中总字节数最少的 flow,从中取一个包发送。
    """
    now = 0.0
    tl = []
    id_counter = itertools.count(1)

    while any(f.has_packets() for f in flows):
        receive_arrivals(flows, now, bufsize, id_counter)
        progressed = False

        candidates = [f for f in flows if f.queue]
        if candidates:
            min_flow = min(candidates, key=lambda f: sum(pkt[2] for pkt in f.queue))
            pid, t_arr, size = min_flow.queue.pop(0)
            tx = size / cap
            now += tx
            min_flow.sent_pkts += 1
            min_flow.sent_bytes += size
            min_flow.wait_sum += now - t_arr
            tl.append((min_flow.name, size, round(now, 4)))
            receive_arrivals(flows, now, bufsize, id_counter)
            progressed = True

        if not progressed:
            na = next_arrival_time(flows)
            if na is not None and na > now:
                now = na
            else:
                now += idle
    return tl


def compute_system_avg_delay(flows):
    """计算系统平均时延"""
    total_wait = sum(f.wait_sum for f in flows)
    total_pkts = sum(f.sent_pkts for f in flows)
    return total_wait / total_pkts if total_pkts > 0 else 0.0


def compute_fairness(flows):
    """计算 Jain 公平性指数 (基于 sent_bytes)"""
    bytes_list = [f.sent_bytes for f in flows]
    n = len(bytes_list)
    sum_x = sum(bytes_list)
    sum_x2 = sum(x**2 for x in bytes_list)
    if sum_x2 == 0:
        return 0.0
    return (sum_x ** 2) / (n * sum_x2)


def compute_drop_rate(flows):
    """计算整体丢包率"""
    total_sent = sum(f.sent_pkts for f in flows)
    total_drop = sum(f.dropped for f in flows)
    total = total_sent + total_drop
    return total_drop / total if total > 0 else 0.0


def print_metrics(name, flows):
    """统一打印所有指标"""
    avg_delay = compute_system_avg_delay(flows)
    fairness = compute_fairness(flows)
    drop_rate = compute_drop_rate(flows)

    print(f"\n--- {name} Metrics ---")
    print(f"System Avg Delay: {avg_delay:.4f}")
    print(f"Jain Fairness Index: {fairness:.4f}")
    print(f"Overall Drop Rate: {drop_rate:.4f}")


def make_flow(name, n, lo, hi, gap):
    sizes = [random.randint(lo, hi) for _ in range(n)]
    arrivals = [i * gap for i in range(n)]
    return Flow(name, sizes, arrivals)


def run_experiments(num_runs=20):
    """运行多次实验,对比 RR / FCFS / SBF 三种算法"""

    scenarios = {
        "Light_Load": {
            "num_pkts": 8,
            "bufsize": 5,
            "flows_config": [
                ("A", 200, 800, 0.5),
                ("B", 300, 700, 0.7),
                ("C", 250, 900, 0.9),
            ]
        },
        "Heavy_Load": {
            "num_pkts": 15,
            "bufsize": 5,
            "flows_config": [
                ("A", 300, 900, 0.3),
                ("B", 400, 800, 0.35),
                ("C", 350, 950, 0.4),
            ]
        }
    }

    results = {scenario: {algo: {"delay": [], "fairness": [], "drop": []}
                         for algo in ["RR", "FCFS", "SBF"]}
               for scenario in scenarios}

    for scenario_name, config in scenarios.items():
        print(f"\n{'='*70}")
        print(f"Running Scenario: {scenario_name}")
        print(f"{'='*70}")

        for seed in range(num_runs):
            random.seed(seed)

            def make_flows():
                return [
                    make_flow(name, config["num_pkts"], lo, hi, gap)
                    for name, lo, hi, gap in config["flows_config"]
                ]

            flows_rr = make_flows()
            flows_fc = make_flows()
            flows_sbf = make_flows()

            round_robin(flows_rr, bufsize=config["bufsize"], cap=1000)
            fcfs(flows_fc, bufsize=config["bufsize"], cap=1000)
            shortest_backlog_first(flows_sbf, bufsize=config["bufsize"], cap=1000)

            for algo_name, flows in [("RR", flows_rr), ("FCFS", flows_fc), ("SBF", flows_sbf)]:
                results[scenario_name][algo_name]["delay"].append(compute_system_avg_delay(flows))
                results[scenario_name][algo_name]["fairness"].append(compute_fairness(flows))
                results[scenario_name][algo_name]["drop"].append(compute_drop_rate(flows))

    print(f"\n{'='*70}")
    print(f"FINAL RESULTS (Average over {num_runs} runs)")
    print(f"{'='*70}\n")

    for scenario_name in scenarios:
        print(f"\n--- {scenario_name} ---")
        print(f"{'Algorithm':<10} {'Avg Delay':<12} {'Fairness':<12} {'Drop Rate':<12}")
        print("-" * 50)

        for algo in ["RR", "FCFS", "SBF"]:
            avg_delay = sum(results[scenario_name][algo]["delay"]) / num_runs
            avg_fairness = sum(results[scenario_name][algo]["fairness"]) / num_runs
            avg_drop = sum(results[scenario_name][algo]["drop"]) / num_runs
            print(f"{algo:<10} {avg_delay:<12.4f} {avg_fairness:<12.4f} {avg_drop:<12.4f}")

    with open("results.csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Scenario", "Algorithm", "Avg_Delay", "Fairness", "Drop_Rate"])
        for scenario_name in scenarios:
            for algo in ["RR", "FCFS", "SBF"]:
                avg_delay = sum(results[scenario_name][algo]["delay"]) / num_runs
                avg_fairness = sum(results[scenario_name][algo]["fairness"]) / num_runs
                avg_drop = sum(results[scenario_name][algo]["drop"]) / num_runs
                writer.writerow([scenario_name, algo, avg_delay, avg_fairness, avg_drop])

    print("\nResults exported to results.csv")


if __name__ == "__main__":
    run_experiments(num_runs=20)
```
